```python
import math
import numpy as np

import matplotlib.pyplot as plt

X_0 = 100
sig = 0.24
mu = 0.11

def pdf(x, t):
    factor1 = 1/x * 1/(sig*math.sqrt(2*math.pi*t))
    
    exp_nominator = -(math.log(x)-math.log(X_0)-(mu-((sig*sig)/2))*t)**2
    exp_denominator = 2*sig*sig*t
    
    factor2 = math.exp(exp_nominator/exp_denominator)
    
    return factor1*factor2


# 1 year:
x = np.arange(1, 200, 1)
t = np.arange(1, 200, 1)

# rows: time, columns: val
matrix = np.zeros((len(x),len(t)))

for index1 in range(0, len(x)):
    for index2 in range(0, len(t)):
        matrix[index1][index2] = pdf(x[index1], t[index2])


X, T = np.meshgrid(x,t)       


from mpl_toolkits.mplot3d import Axes3D
    
fig = plt.figure(figsize=(15, 10), dpi=80)
ax = Axes3D(fig)
ax.plot_surface(X,T,matrix)
ax.set_title('Probability density function')
plt.xlabel('Price')
plt.ylabel('Time')
plt.show()
```
